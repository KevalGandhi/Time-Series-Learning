```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
possum = pd.read_csv('possum.csv')

possum = pd.get_dummies(possum)

possum = possum.dropna()

#this tells you if any are NaN and in which colummns
possum.isnull().any()
```

```python
possum.info()
possum.describe()
possum.columns

possum.drop('case', axis = 'columns', inplace = True)
print(possum.corr())
```

```python
covariate = possum[['age','skullw','totlngth','taill',
       'footlgth', 'earconch','eye', 'chest', 'Pop_Vic', 'Pop_other',
       'sex_f', 'sex_m']]

variate = possum[['hdlngth']]

print(covariate.corr())

#by looking at corr matrix we get rid of site, belly, pop_vic and pop_other
'''
[['hdlngth', 'skullw', 'totlngth','taill',
       'footlgth', 'earconch','eye', 'chest', 'Pop_Vic', 'Pop_other',
       'sex_f', 'sex_m']]
'''
```

```python
def performmodel(iterations, lamb):
    totalr = 0
    totalmse = 0
    for i in range(0,iterations):
        x_train, x_test, y_train, y_test = train_test_split(covariate, variate, test_size = 0.3)
        
        '''
        print(x_train)
        print(x_test)
        print(y_train)
        print(y_test)
        '''

        #linear regression
        '''
        model_fit = LinearRegression().fit(x_train, y_train)
        print(model_fit.coef_, model_fit.intercept_)
        '''
        #ridge regression
        '''
        model_fit = Ridge(alpha=lamb).fit(x_train, y_train)
        '''
        
        #lasso regression
        model_fit = Lasso(alpha=lamb).fit(x_train, y_train)
        #print(model_fit.coef_, model_fit.intercept_)
        
        y_pred = model_fit.predict(x_test)
        totalr += model_fit.score(x_test, y_test)
        totalmse += mean_squared_error(y_test, y_pred)

    print(totalr/iterations, np.sqrt(totalmse/iterations))

performmodel(1000, 0)
```
